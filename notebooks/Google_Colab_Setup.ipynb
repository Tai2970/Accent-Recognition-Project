```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

AUDIO_FILE_URL = "https://raw.githubusercontent.com/Tai2970/Accent-Recognition-Project/james/data/test.wav"
LOCAL_FILENAME = "data_test.wav"

!wget {AUDIO_FILE_URL} -O {LOCAL_FILENAME}

# load the audio file
y, sr = librosa.load(LOCAL_FILENAME, sr=None)  # sr=None keeps the original sample rate

# print basic audio properties
duration = librosa.get_duration(y=y, sr=sr)
print(f"Sample Rate: {sr}")
print(f"Duration: {duration:.2f} seconds")

# plot the waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
