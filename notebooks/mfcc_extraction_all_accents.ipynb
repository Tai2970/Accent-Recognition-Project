```python
import librosa
import soundfile as sf
import os

input_folder = "data/"
target_sample_rate = 16000

for accent in ["american", "british", "vietnamese"]:
    accent_input = os.path.join(input_folder, accent)
    accent_output = os.path.join(accent_input, "processed")

    os.makedirs(accent_output, exist_ok=True)

    for filename in os.listdir(accent_input):
        if filename.endswith(".wav") and not filename.startswith("processed_"):
            file_path = os.path.join(accent_input, filename)
            y, sr = librosa.load(file_path, sr=None)
            y_resampled = librosa.resample(y, orig_sr=sr, target_sr=target_sample_rate)
            output_file_path = os.path.join(accent_output, f"processed_{filename}")
            sf.write(output_file_path, y_resampled, target_sample_rate)
            print(f"Processed: {filename} -> {accent}/processed/")

print("All files have been converted to 16kHz and saved in their respective accent folders.")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

data_folders = {
    "American": "data/american/processed/",
    "British": "data/british/processed/",
    "Vietnamese": "data/vietnamese/processed/"
}

for accent, folder in data_folders.items():
    audio_files = [f for f in os.listdir(folder) if f.endswith(".wav")]

    if len(audio_files) == 0:
        print(f"No files found in {folder}, skipping...")
        continue

    for file_name in audio_files:  # Process each file
        file_path = os.path.join(folder, file_name)
        y, sr = librosa.load(file_path, sr=16000)
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

        print(f"{accent} - {file_name} MFCCs Shape: {mfccs.shape}")

        plt.figure(figsize=(10, 4))
        librosa.display.specshow(mfccs, x_axis="time", sr=sr)
        plt.colorbar(label="MFCC Coefficients")
        plt.title(f"MFCCs of {accent} - {file_name}")
        plt.xlabel("Time (s)")
        plt.ylabel("MFCC Coefficients")
        plt.show()
```

```python
import os
import numpy as np
import librosa

# Define accents
accents = ["american", "british", "vietnamese"]

# Process each accent's processed audio files and extract MFCCs
for accent in accents:
    input_folder = f"data/{accent}/processed/"
    output_folder = f"data/{accent}/mfcc_features/"

    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            file_path = os.path.join(input_folder, filename)

            # Load audio
            y, sr = librosa.load(file_path, sr=16000)

            # Extract MFCCs
            mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

            # Save MFCCs as .npy
            np.save(os.path.join(output_folder, filename.replace(".wav", ".npy")), mfccs)

            print(f"Extracted MFCCs: {filename} ({accent.capitalize()})")

print("All MFCC features have been extracted and saved in 'mfcc_features/' folders for all accents.")

# List all MFCC files per accent
for accent in accents:
    mfcc_folder = f"data/{accent}/mfcc_features/"
    print(f"\n{accent.capitalize()} MFCC Files:")
    print(os.listdir(mfcc_folder))

# Cleanup: Remove unprocessed MFCC files that don't start with "processed_"
for accent in accents:
    mfcc_folder = f"data/{accent}/mfcc_features/"

    for filename in os.listdir(mfcc_folder):
        if filename.startswith("common_voice_en_") and filename.endswith(".npy"):
            file_path = os.path.join(mfcc_folder, filename)
            os.remove(file_path)
            print(f"Deleted: {filename} ({accent.capitalize()})")

print("Cleanup complete! Only processed MFCC files remain.")

# List final MFCC files after cleanup
for accent in accents:
    mfcc_folder = f"data/{accent}/mfcc_features/"
    print(f"\nFinal {accent.capitalize()} MFCC Files After Cleanup:")
    print(os.listdir(mfcc_folder))
```
